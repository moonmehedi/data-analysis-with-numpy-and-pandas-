```python
retail.loc[:,['sales','onpromotion']].sample(100,random_state=12).mean().round(2)
```

GROUP BY METHOD

```python
#returns  a dataframe 
retail.groupby('family')[['sales']].sum()

```

```python
retail.groupby(['family','store_nbr'])[['sales']].sum()
```

MULTIPLE AGGREGATION

```python
retail.groupby(['family','store_nbr'])[['sales','onpromotion']].agg(['sum','mean'])
```

```python
#aggregation on entire columns

retail.drop(['date'],axis=1).groupby(['family','store_nbr']).agg(['sum','mean'])
```

```python
retail=retail.drop(['date'],axis=1)
retail
```

```python
#different aggregation based on column

retail.groupby(['family','store_nbr']).agg({'sales':['sum','mean','std'],
                                                'onpromotion':['min','max']})

#id is not present as i have not mentioned it
```

PIVOT TABLE


```python
retail.pivot_table(
    index='family',
    columns='store_nbr',
    values='sales',
    aggfunc='sum',
    margins='True'
)
```

MULTIPLE AGGREGATION FUNCTION IN A PIVOT TABLE


```python
retail
```

```python
retail.pivot_table(
    index='family',
    columns=('store_nbr'),
    values='sales',
    aggfunc=('min','max')
)
```

```python
retail.pivot_table(
    index='family',
    columns=('store_nbr'),
    aggfunc=({'sales':['sum','mean'],"onpromotion":'max'})
)
```

HEATMAP

```python
retail.pivot_table(
    index='family',
    columns='store_nbr',
    aggfunc=({'sales':'sum'})
).style.background_gradient(cmap='RdYlGn',axis=None)
```
